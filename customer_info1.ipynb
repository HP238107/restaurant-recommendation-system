```python
import pandas as pd
import numpy as np
```

```python
train_locations=pd.read_csv("C:/Users/hp/Downloads/case study -1/train_locations.csv")
train_customers=pd.read_csv("C:/Users/hp/Downloads/case study -1/train_customers.csv")
```

```python
train_customers.rename(columns={'akeed_customer_id':'customer_id'},inplace=True)
train_customers.head()
```

```python
train_locations['location_type'].fillna('Other',inplace=True)

train_locations.head()
```

```python
df1= train_customers.loc[0: ,['customer_id','gender']]
df2= train_locations.loc[0: ,['customer_id','location_type','longitude','latitude']]
```

```python
df2.dropna(inplace=True)
```

```python
df1 =df1.sort_values(by=['customer_id'])
df1.drop_duplicates(subset='customer_id',keep='last',inplace=True)
df1.head()
```

```python
df2 =df2.sort_values(by=['customer_id'])
df2.drop_duplicates(subset='customer_id',keep='last',inplace=True)
df2.head()
```

```python
customers_info=pd.merge(df1,df2,how="right",left_on='akeed_customer_id',right_on='customer_id')
del customers_info['akeed_customer_id']
```

```python
consumer=pd.merge(df1,df2,on='customer_id')
consumer.head()
```

```python
customers_info=customers_info.rename(columns={'latitude':'cust_latitude','longitude':'cust_longitude'})
customers_info.head()
```

```python
def gender(x):
    '''
    values in gender were of class "Male","Male  ","male " so this function brings them to base
    '''
    try:
      x1=x.split(" ")[0]
      x1=x1.lower()
      return x1
    except:
      return np.nan
consumer.gender=consumer.gender.apply(gender)
```

```python
consumer['gender'] = consumer['gender'].fillna('female')
```
